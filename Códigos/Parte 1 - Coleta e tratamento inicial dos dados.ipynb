#### Passo 1 - Fazer upload dos dados de demanda e juntar em um dataframe

```python
df_1 = pd.read_csv('Demanda\Demanda_2020.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5)   
df_2 = pd.read_csv('Demanda\Demanda_2019.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_3 = pd.read_csv('Demanda\Demanda_2018.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_4 = pd.read_csv('Demanda\Demanda_2017.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_5 = pd.read_csv('Demanda\Demanda_2016.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_6 = pd.read_csv('Demanda\Demanda_2015.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_7 = pd.read_csv('Demanda\Demanda_2014.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_8 = pd.read_csv('Demanda\Demanda_2013.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_9 = pd.read_csv('Demanda\Demanda_2012.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_10 = pd.read_csv('Demanda\Demanda_2011.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_11 = pd.read_csv('Demanda\Demanda_2010.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_12 = pd.read_csv('Demanda\Demanda_2009.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_13 = pd.read_csv('Demanda\Demanda_2008.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_14 = pd.read_csv('Demanda\Demanda_2007.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5) 
df_15 = pd.read_csv('Demanda\Demanda_2006.csv', sep=';', header=None, names=['Horário','Data', 'Demanda', 'Unidade'],skiprows=5)
```

```python
df= pd.concat([df_1, df_2, df_3,df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11, df_12, df_13, df_14, df_15],ignore_index=True)
df.shape
```

```python
df.head()
```

Deixar apenas a data e horário antes de realizar a transformação na coluna Data

```python
df['Horário'] = df['Horário'].str[-8:]
df['Data'] = df['Data'].str[:10]
df['Demanda'] = df['Demanda'].str[:-3]
df.head()
```

```python
df['Horário'] = pd.to_datetime(df['Horário'], format = '%H:%M:%S', errors='coerce')
df['Data'] = pd.to_datetime(df['Data'], format = '%d/%m/%Y', errors='coerce')
df['Demanda'] = pd.to_numeric(df['Demanda'], errors='coerce')
df.head()
```

Realizando alguns ajustes no dataframe:

```python
#Multiplicando a demanda por 1000
df['Demanda'] = df['Demanda']*1000
#Transformando o horário em tipo tempo
df['Time'] = df['Horário'].dt.time
#Incluir dia da semana
df['Dia da Semana'] = df['Data'].dt.weekday
df.tail()
```

```python
### Excluir colunas desnecessárias
df = df.drop(columns=['Horário','Unidade'])
df = df.rename(columns={"Time": "Horário"})
df.head()
```

Verificação de valores não existentes

```python
df.isna().sum()
```

Como existem poucos valores ausentes, eles serão excluídos.

```python
df_cleaned = df.dropna()
len(df_cleaned)
```

#### Passo 2 - Importando dados climáticos

```python
sul = pd.read_csv('OutrosDados\DadosclimaBage.csv', sep=";", index_col=False)
metro = pd.read_csv('OutrosDados\DadosclimaPortoAlegre.csv', sep=";", index_col=False)
centro = pd.read_csv('OutrosDados\DadosclimaSantaMaria.csv', sep=";", index_col=False)
serra = pd.read_csv('OutrosDados\DadosclimaCaxias.csv', sep=";", index_col=False)
sul.head()
```

Transformar a coluna "Hora" em horário.

```python
sul.loc[sul['Hora'] == 0, 'horário'] = '00:00:00'  
sul.loc[sul['Hora'] == 1200, 'horário'] = '12:00:00'  
sul.loc[sul['Hora'] == 1800, 'horário'] = '18:00:00'
metro.loc[metro['Hora'] == 0, 'horário'] = '00:00:00'  
metro.loc[metro['Hora'] == 1200, 'horário'] = '12:00:00'  
metro.loc[metro['Hora'] == 1800, 'horário'] = '18:00:00'
centro.loc[centro['Hora'] == 0, 'horário'] = '00:00:00'  
centro.loc[centro['Hora'] == 1200, 'horário'] = '12:00:00'  
centro.loc[centro['Hora'] == 1800, 'horário'] = '18:00:00'
serra.loc[serra['Hora'] == 0, 'horário'] = '00:00:00'  
serra.loc[serra['Hora'] == 1200, 'horário'] = '12:00:00'  
serra.loc[serra['Hora'] == 1800, 'horário'] = '18:00:00'
```

```python
sul['Data'] = pd.to_datetime(sul['Data'], format = '%d/%m/%Y', errors='coerce')
sul['Horário'] = pd.to_datetime(sul['horário'], format = '%H:%M:%S', errors='coerce').dt.time
metro['Data'] = pd.to_datetime(metro['Data'], format = '%d/%m/%Y', errors='coerce')
metro['Horário'] = pd.to_datetime(metro['horário'], format = '%H:%M:%S', errors='coerce').dt.time
centro['Data'] = pd.to_datetime(centro['Data'], format = '%d/%m/%Y', errors='coerce')
centro['Horário'] = pd.to_datetime(centro['horário'], format = '%H:%M:%S', errors='coerce').dt.time
serra['Data'] = pd.to_datetime(serra['Data'], format = '%d/%m/%Y', errors='coerce')
serra['Horário'] = pd.to_datetime(serra['horário'], format = '%H:%M:%S', errors='coerce').dt.time
sul.head()
```

Excluir as colunas que não serão utilizadas

```python
sul = sul.drop(columns=['Estacao','TempBulboUmido','Hora','PressaoAtmEstacao', 'DirecaoVento','horário'])
metro = metro.drop(columns=['Estacao','TempBulboUmido','Hora','PressaoAtmEstacao', 'DirecaoVento','horário'])
centro = centro.drop(columns=['Estacao','Hora','TempBulboUmido','PressaoAtmEstacao', 'DirecaoVento','horário'])
serra = serra.drop(columns=['Estacao','Hora','TempBulboUmido', 'PressaoAtmEstacao', 'DirecaoVento','horário'])
serra.head()
```

Renomear colunas para que conste a região do estado e juntar todos os dataframes em apenas um dataframe.

```python
sul = sul.rename(columns={"TempBulboSeco": "Sul_TempSeco", "TempBulboUmido": "Sul_TempUmido", "UmidadeRelativa":"Sul_Umidade",
                                 "VelocidadeVento": "Sul_Vento", "Nebulosidade": "Sul_Nebulosidade"})
metro = metro.rename(columns={"TempBulboSeco": "Metro_TempSeco", "TempBulboUmido": "Metro_TempUmido", "UmidadeRelativa":"Metro_Umidade",
                                 "VelocidadeVento": "Metro_Vento", "Nebulosidade": "Metro_Nebulosidade"})
centro = centro.rename(columns={"TempBulboSeco": "Centro_TempSeco", "TempBulboUmido": "Centro_TempUmido", "UmidadeRelativa":"Centro_Umidade",
                                 "VelocidadeVento": "Centro_Vento", "Nebulosidade": "Centro_Nebulosidade"})
serra = serra.rename(columns={"TempBulboSeco": "Serra_TempSeco", "TempBulboUmido": "Serra_TempUmido", "UmidadeRelativa":"Serra_Umidade",
                                 "VelocidadeVento": "Serra_Vento", "Nebulosidade": "Serra_Nebulosidade"})
```

```python
df_clima = sul
df_clima = df_clima.merge(metro, how='left', on = ['Horário','Data'])
df_clima = df_clima.merge(centro, how='left', on = ['Horário','Data'])
df_clima = df_clima.merge(serra, how='left', on = ['Horário','Data'])
df_clima.head()
```

```python
df_clima['Temperatura'] = (df_clima['Sul_TempSeco']+df_clima['Metro_TempSeco']+df_clima['Serra_TempSeco']+df_clima['Centro_TempSeco'])/4
df_clima['Umidade'] = (df_clima['Sul_Umidade']+df_clima['Metro_Umidade']+df_clima['Serra_Umidade']+df_clima['Centro_Umidade'])/4
df_clima['Vento'] = (df_clima['Sul_Vento']+df_clima['Metro_Vento']+df_clima['Serra_Vento']+df_clima['Centro_Vento'])/4
df_clima['Nebulosidade'] = (df_clima['Sul_Nebulosidade']+df_clima['Metro_Nebulosidade']+df_clima['Serra_Nebulosidade']+df_clima['Centro_Nebulosidade'])/4
```

```python
df_clima = df_clima.drop(columns=['Sul_TempSeco', 'Metro_TempSeco', 'Serra_TempSeco', 'Centro_TempSeco','Sul_Umidade',
                        'Metro_Umidade', 'Serra_Umidade', 'Centro_Umidade', 'Sul_Vento', 'Metro_Vento', 'Serra_Vento',
                        'Centro_Vento','Sul_Nebulosidade','Metro_Nebulosidade','Serra_Nebulosidade','Centro_Nebulosidade'])
df_clima.head()
```

Juntando o dataframe de clima com o de demanda

```python
df_merged = df_cleaned.merge(df_clima, how='left', on = ['Horário','Data'])
df_merged.head()
```

```python
df_merged = df_merged.sort_values(by=['Data', 'Horário'], ascending=True).reset_index(drop=True)
```

Como os dados climáticos eram para apenas  horários do dia (00:00, 12:00 e 18:00) foi realizada uma interpolação para que todos os minutos do dia tivessem o dado climático. 

```python
df_merged.interpolate(method='linear',axis=0, inplace=True)
df_merged.head()
```

#### Passo 3 -Importando índice de atividade econômica do Brasil

```python
iaebr = pd.read_csv('OutrosDados\IAEBR.csv', sep=";", index_col=False)
iaebr.head()
```

Corrigindo os tipos e formatos dos elementos do dataframe

```python
iaebr = iaebr.replace(to_replace=',', value='.', regex=True)
iaebr['data'] = pd.to_datetime(iaebr['data'], format = '%d/%m/%Y', errors='coerce')
iaebr['valor'] = pd.to_numeric(iaebr['valor'])
iaebr.dtypes
```

Juntando com os dados de demanda

```python
df_merged['Mês'] = df_merged['Data'].dt.to_period(freq='M')
iaebr['Mês'] = iaebr['data'].dt.to_period(freq='M')
iaebr = iaebr.drop(columns=['data'])
df_merged = df_merged.merge(iaebr, how='left', on = 'Mês')
df_merged.head()
```

#### Passo 4 - Importando índice de Desempenho Industrial do RS

```python
desempenho_industrial = pd.read_excel('OutrosDados\Desempenho_Industrial.xlsx', parse_dates=True)
desempenho_industrial.head()
```

Juntando ao dataframe anterior

```python
desempenho_industrial['Mês'] = desempenho_industrial['Unnamed: 0'].dt.to_period(freq='M')
desempenho_industrial = desempenho_industrial.drop(columns=['Unnamed: 0'])
df_merged = df_merged.merge(desempenho_industrial, how='left', on = 'Mês')
df_merged.head()
```

#### Passo 5 - Importando dados de população

```python
popul = pd.read_excel('OutrosDados\População.xlsx')
popul.head()
```

Juntando ao dataframe

```python
popul['Ano'] = popul['Ano'].dt.to_period(freq='Y')
df_merged['Ano'] = df_merged['Data'].dt.to_period(freq='Y')
df_merged = df_merged.merge(popul, how='left', on = 'Ano')
df_merged.head()
```

Verificando os tipos dos elementos no dataframe final

```python
df_merged.dtypes
```

```python
df_merged['valor'] = pd.to_numeric(df_merged['valor'], errors='coerce')
df_merged = df_merged.drop(columns=['Mês','Ano'])
```

#### Passo 6 - Importando dados de feriados

```python
calend = pd.read_excel('feriados.xlsx')
calend.head()
```

```python
calend = calend.drop(columns=['Dia da Semana'])
calend.tail()
```

Inclusão de feriado estadual

```python
df_RS = pd.concat([pd.DataFrame([str(i)+'-09-20'], columns = ['Data']) for i in range (2007, 2078)],
          ignore_index=True)
df_RS['Feriado']='Revolução Farroupilha'
df_RS.head()
```

```python
calend = pd.concat([calend, df_RS],ignore_index=True)
calend['Data'] = pd.to_datetime(calend['Data'], format = '%Y-%m-%d', errors='coerce')
df_merged = df_merged.merge(calend, how='left', on = 'Data')
```

Incluir coluna em que os dias sem feriado sejam preenchidos com a string "Sem Feriado"

```python
df_merged['Feriado'] = df_merged['Feriado'].fillna('Sem feriado')
df_merged.head()
```
